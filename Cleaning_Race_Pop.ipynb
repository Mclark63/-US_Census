```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
black_data_load = "population_black.csv"

# Read School and Student Data File and store into Pandas DataFrames
race_population_data = pd.read_csv(black_data_load)


race_population_data.head(10)
```

```python
removemargin = race_population_data.drop(race_population_data.filter(regex='Margin').columns, axis=1)
removemargin.head
```
