```python
NDF = population_data[['Label (Grouping)','Georgia!!Total!!Estimate', 'Georgia!!Male!!Estimate', 'Georgia!!Female!!Estimate', ]]

NDF.head(10)
```

```python
NewColums = NDF.rename(columns={'Label (Grouping)': 'Age Group', 'Georgia!!Total!!Estimate': 'Overall Total Est.','Georgia!!Male!!Estimate': 'Est. Males Total', 'Georgia!!Female!!Estimate': 'Est. Females Total'})

NewColums.head()
```

```python
NewColums.shape
```

```python
# Drop all columns

NewColums.dropna(subset=['Overall Total Est.','Est. Males Total','Est. Females Total'], inplace=True)

NewColums.head(42)
```

```python
NewColums['Est. Females Total'] = NewColums['Est. Females Total'].replace('(X)', pd.NA)

# Step 2: Drop rows with NaN values in the specified column
NewColums.dropna(subset=['Est. Females Total'], inplace=True)

NewColums.head(42)
```
